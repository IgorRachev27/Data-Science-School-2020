### Определение функции, области определения и значения
Подробнее: https://drive.google.com/file/d/1dM7c5_yOBBUepEA0Q_5wokKUNG_GiCmt/view

Говорят, что на некотором множестве $D$ имеется **функция $f$ со значениями из множества $E$**, если для каждого элемента $x$ из $D$ по правилу $f$ поставлен в соответствие некоторый элемент y из множества $E$. 

Другими словами, функция $f$ отображает множество $D$ в множество $E$. При этом множество $D$ называется **областью определения**, а множество $E$ — **областью значений** данной функции $f$. 

Возьмём два множества — множество граждан России и множество номеров их российских паспортов. Ясно, что у каждого гражданина имеется свой номер паспорта.


<img src="https://github.com/ddvika/Data-Science-School-2020/blob/main/lecture_8/imgs/set_ex.png?raw=true" width=500 height=350 />

Получаем соответствие, при котором каждому гражданину России сопоставляется определённый набор цифр — номер его паспорта. Это соответствие проиллюстрировано на рисунке.

Более того, номер паспорта уникален: по номеру паспорта можно однозначно найти конкретного человека. Такое соответствие в математике называется **взаимно-однозначным.**

Линейная функция $y = kx + b$ при $k \neq 0$ является примером взаимно-однозначного соответствия. Возьмём, к примеру, функцию $y = 3x + 1$. Каждому значению $x$ здесь соответствует своё значение $y$. И наоборот — каждому $y$ соответствует одно-единственное значение $x$.
$$
\begin{array}{|c|c|c|c|c|c|c|c|c|}
\hline x & -2 & -1 & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline y & -5 & -2 & 1 & 4 & 7 & 10 & 13 & 16 \\
\hline
\end{array}
$$


В свою очередь, **множество значений $Y$ функции** — это множество таких $y \in E$, что существует $x \in D$ такой, что $f(x) = y$. (Множество значений Y функции, вообще говоря, может быть подмножеством области
значений E.)

- Например, областью определения функции $y = sin(x)$ служит множество всех действительных чисел, а областью значений — отрезок $[−1; 1]$. 

- Областью определения функции $y=\sqrt{x-1}$ является промежуток $[1; +∞)$, а областью значений — промежуток $[0; +∞)$.



Наиболее интуитивное представление можно получить из геометрического места точек (ГМТ) (назваемого графиком функции) плоскости таких $y = f(x)$. (Функцию называют непрерывной, если, говоря нестрого, малые изменения её аргумента приводят к малым изменениям её значения. Если это условие нарушается, функция терпит разрыв.)

Есть несколько способов задания числовой функции.
1. С помощью формулы. Это наиболее привычный для нас способ. Примеры: $y = cos(x)$, $y = x^22 − 2x$.
2. Наглядный способ — с помощью графика. Например, зависимость курса доллара от вре- мени удобнее всего смотреть на графике. Вряд ли возможно задать эту зависимость какой- либо формулой :-)
3. С помощью таблицы. Этот способ будет единственно возможным, например, в ситуации, когда снимается экспериментальная зависимость, формула ещё не выведена и график ещё не построен.
4. С помощью описания того, как устроено соответствие. Например, функция $y = |x|$ по сути задаётся описанием:

$$
y=\left\{\begin{aligned}
x, & \text { если } x \geqslant 0 \\
-x, & \text { если } x<0
\end{aligned}\right.
$$



### Более простым языком о функциях
**Функция (отображеение, оператор, преобразоваание)** — в математике соответствие между элементами двух множеств, установленное по такому правилу, что каждому элементу первого множества соответствует один и только один элемент второго множества.

**Область определения**  — множество, на котором задаётся функция. В каждой точке этого множества значение функции должно быть определено.

**Область значений (или множество значений) функции** — множество, состоящее из всех значений, которые принимает функция.

### Примеры

1. Найти область определения функции 
$$
f(x)=\frac{x+2}{x^{2}-3}
$$

Решение
в числителе ничего особенного нет, а вот знаменатель должен быть ненулевым. Давайте приравняем его к нулю и попытаемся найти «плохие» точки:
$$
\begin{array}{l}
x^{2}-3=0 \\
x^{2}=3
\end{array}
$$

Полученное уравнение имеет два корня: $x=-\sqrt{3}, x=\sqrt{3}$. Данные значения не входят в область определения функции. Действительно, подставьте $x=-\sqrt{3}$ или $x=\sqrt{3} $в функцию и вы увидите, что знаменатель обращается в ноль.

Ответ: область определения: $D(f)=R \backslash\{-\sqrt{3} ; \sqrt{3}\}$

Вопрос: что означает запись $R \backslash X$?

2. Найти область определения функции 
$$
f(x)=-\frac{2}{4-x^{2}}
$$

3. Найти область определения функции 

$$
f(x)=\frac{2 x+3}{x^{2}+2 x+5}
$$

4. Найти область значений функции $$
f(x)=x+\frac{1}{x}
$$

Решение. Искомая область значений есть множество всех a, при которых уравнение
$$
x+\frac{1}{x}=a
$$
имеет решение. Преобразуем:
$$
\frac{x^{2}-a x+1}{x}=0 \quad \Leftrightarrow \quad x^{2}-a x+1=0
$$
Полученное квадратное уравнение имеет корни при неотрицательном дискриминанте:
$$
D=a^{2}-4 \geqslant 0
$$
откуда $ a \leqslant-2$ или $a \geqslant 2$.

Ответ: $E(f)=(-\infty ;-2] \cup[2 ;+\infty)$

5. Найти наибольшее и наименьшее значение функции $$
f(x)=\frac{x}{x^{2}-x+1}
$$
Решение. Давайте просто найдём область значений данной функции. Ищем все значения a, при
 которых уравнение
 $$
\frac{x}{x^{2}-x+1}=a
$$
имеет решения. Умножаем обе части на выражение $x^2 − x + 1$, которое не обращается в нуль ни при каком $x$, и после преобразований получаем:
$$
a x^{2}-(a+1) x+a=0
$$
Если $a = 0$, то уравнение (1) имеет корень x = 0, так что a = 0 годится.
Если $a \neq 0$, то уравнение (1) является квадратным. Чтобы оно имело корни, его дискриминант должен быть неотрицателен:
$$
D=-3 a^{2}+2 a+1 \geqslant 0
$$

откуда $-\frac{1}{3} \leqslant a \leqslant 1$ Этот отрезок содержит значение a = 0, полученное ранее. 

Итак, мы нашли область значений $E(f)=\left[-\frac{1}{3} ; 1\right]$ Теперь ясно, что наибольшее значение функции $f$ равно 1, а наименьшее значение равно $-\frac{1}{3}$.

Ответ: 1 и $-\frac{1}{3}$.

### Формальное определение производной

Подробнее: https://drive.google.com/file/d/1dM7c5_yOBBUepEA0Q_5wokKUNG_GiCmt/view,

http://mathprofi.ru/opredelenie_proizvodnoi_smysl_proizvodnoi.html

Символом $\mathbb{R}$ будет обозначено множество вещественных (или действительных) чисел.
Предположим, что у нас есть функция $f: \mathbb{R} \rightarrow \mathbb{R}$, вход и выход которой являются скалярами. Производная от $f$ определяется как

$$f'(x) = \lim_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h},$$

если этот предел существует. Если $f'$ существует, говорят что $f$ называется дифференцируемой в $x$. Если $f$ дифференцируема в каждой точке интервала, то эта функция дифференцируема на этом интервале. Мы можем интерпретировать производную $f'(x)$ как мгновенную скорость изменения $f(x)$ по отношению к $x$. Так называемая мгновенная скорость изменения основана на изменении $h$ в $x$, которое приближается к 0.

### Вывод определения

Рассмотрим это формальное определение производной и поймем откуда оно происходит. Для этого нам понадобится ввести несколько понятий.

Пусть задана некоторая функци $f(x) = (1+x)^{\frac{1}{x}}$. Эта функция не определена в точке x = 0, но ее значение может быть вычислено в точках, сколь угодно
близких к точке 0. Другими словами, будет изучаться ее поведение при $x \to 0$. Напишем функцию, которая расчитывает $f(x)$ в зависимости от $х$.

```python
def compute_f_x(x):
    return (1 + x)**(1 / x)

x = 0.1
for i in range(10):
    # рассчитываем по обозначенной формуле значение
    print(f'Значение х = {x}, f(x) = {compute_f_x(x)}')
    # на каждой итерации делим на 10
    x *= 0.1
```

### Понятие предела

**Предел фуункции (предельное значение функции) в заданной точке**, предельной для области определения функции, — такая величина, к которой стремится значение рассматриваемой функции при стремлении её аргумента к данной точке. 

**Пример.** Найти предел функции $f(x)=\frac{x^{2}+x-1}{x-1}$ при $x \rightarrow 0$

**Решение.** Подставляем вместо x значение 0. Получаем:
$$
\lim _{x \rightarrow 0} \frac{x^{2}+x-1}{x-1}=\frac{0+0-1}{0-1}=\frac{-1}{-1}=1
$$
Итак, предел данной функции при  $x \rightarrow 0$ равен 1.


Можно заметить, что чем меньше значение х, тем ближе мы приближаемся к некоторой величине. По этой причине вводят обозначение, называемое пределом:

$$\lim_{x \to 0} (1+x)^{\frac{1}{x}} = e$$

Однако не все функции имеют конечный предел, например, для $g(x) = \frac {1}{x}$ значения на выходе будут возрастать в 10 раз. Другими словами функция будет неограниченно расти при приближении к 0:

```python
def compute_g_x(x):
    return (1/x)

x = 0.1
for i in range(10):
    # рассчитываем по обозначенной формуле значение
    print(f'Значение х = {x}, f(x) = {compute_g_x(x)}')
    # на каждой итерации делим на 10
    x *= 0.1
```

### Свойства предела:

1.  $$
\lim _{x \rightarrow x_{0}}[f(x) \pm g(x)]=\lim _{x \rightarrow x_{0}} f(x) \pm \lim _{x \rightarrow x_{0}} g(x)
$$
2. $$
\lim _{x \rightarrow x_{0}}[f(x) \cdot g(x)]=\lim _{x \rightarrow x_{0}} f(x) \cdot \lim _{x \rightarrow x_{0}} g(x)

## Непрерывность и разрывы

### Непрерывность

Говорят, что функция является непрерывной в точке $x = a$, если

$$\lim_{x \to a} f(x) = f(\lim_{x \to a} x) = f(a)$$

Иными словами: 

**Непрерывная функция** — функция, которая меняется без мгновенных «скачков», то есть такая, малые изменения аргумента которой приводят к малым изменениям значения функции. График непрерывной функции является непрерывной линией

### Разрывы и их типы



Особо стоит выделить так называемые функции с устранимым разрывом в некоторой точке. В таком случае переопределение значения функции в этой точке приводит к тому, что функция становится непрерывной:

![ustranim](https://github.com/tixonsit/mmdad_materials/blob/master/ustranim.png?raw=true)

Графики функций с уранимым разрывом в точке $x_0$ и с разрывами в точках $a$, $b$

Точка устранимого разрыва: Если существуют левый и правый пределы функции в точке и они равны друг другу, но не совпадают со значением функции f(x) в точке $a: f(a)≠f(a−0)=f(a+0)$ или функция $f(x)$ не определена в точке a, то точка a называется точкой **устранимого разрыва.**


Другим важным случаем является разрыв типа «скачок». Также встречаются случаи, когда «функция
уходит в бесконечность» в точке разрыва:

### Вывод производной

С помощью понятия предела определяется другое полезное понятие — понятие производной. Но прежде всего следует рассмотреть линейную функцию. Для нее легко можно получить связь коэффициента наклона и скорости роста функции (Пример геометрического построения будет рассмотрен далее):

$$k = \frac{f(x + h) - f(x)}{h}$$

Откуда, имея определение предела получаем знакомую формулу для производной:

$$f'(x) = \lim_{h \rightarrow 0} \frac{f(x+h) - f(x)}{h},$$

Гладкой называется такая функция, производная которой существует и непрерывна.

### Секущая и касательная

Секущей графика называется прямая, проходящая через две точки графика. Пусть $f(x)$ — некоторая функция. Секущая, которая будет пересекать график в точках, соответствующих значениям аргумента $x_0$ и $x_0 + \Delta x$, является единственной и имеет уравнение:

$$y = f(x_0) + \frac {\Delta y}{\Delta x} (x - x_0)$$

Тут $\Delta y = f(x_0 +\Delta x) - f(x_0)$

![dsbig](https://github.com/tixonsit/mmdad_materials/blob/master/dsbig.png?raw=true)

Если теперь при зафиксированном $x_0$ перейти к пределу $\Delta x \to 0$, уравнение секущей перейдет в уравнение
касательной. Действительно, уравнение касательной имеет вид:

$$y = f(x_0) + f'(x_0)(x-x_0) = f(x_0) + \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x} (x - x_0)$$

Доп:
Пусть, например, в точке x = 0 производная испытывает разрыв. Это соответствует тому, что угол наклона касательной тоже испытывает скачок при переходе через эту точку. Появляется «излом». Другими словами, гладкие функции не могут иметь изломов, поскольку производная непрерывна. Это очень хорошо соответствует интуитивным представлениям о гладких функциях.

### Вычисление численно и правила дифференцирования

Давайте познакомимся с несколькими эквивалентными обозначениями для производных. Дано $y = f(x)$, где $x$ и $y$ - независимая переменная и зависимая переменная функции $f$ соответственно. Следующие выражения эквивалентны:

$$f'(x) = y' = \frac{dy}{dx} = \frac{df}{dx} = \frac{d}{dx} f(x) = Df(x) ,$$

где символы $\frac{d}{dx}$ и $D$ являются операторами дифференцирования, которые указывают на операцию дифференцирования. Мы можем использовать следующие правила для дифференциации общих функций:

* $DC = 0$ ($C$ некоторая константа),
* $Dx^n = nx^{n-1}$ (степенное, где $n$ любое вещественное число),
* $De^x = e^x$,
* $D\ln(x) = 1/x.$

Чтобы отличить функцию, которая образована от нескольких более простых функций, таких как перечисленные выше общие функции, нам могут пригодиться следующие правила. Предположим, что функции $f$ и $g$ дифференцируемы, а $C$ - некая константа, у нас есть правила дифференцирования:

$$\frac{d}{dx} [Cf(x)] = C \frac{d}{dx} f(x),$$

Для суммы:

$$\frac{d}{dx} [f(x) + g(x)] = \frac{d}{dx} f(x) + \frac{d}{dx} g(x),$$

Для произведения:

$$\frac{d}{dx} [f(x)g(x)] = f(x) \frac{d}{dx} [g(x)] + g(x) \frac{d}{dx} [f(x)],$$

Для частного

$$\frac{d}{dx} \left[\frac{f(x)}{g(x)}\right] = \frac{g(x) \frac{d}{dx} [f(x)] - f(x) \frac{d}{dx} [g(x)]}{[g(x)]^2}.$$

Для сложной функции

$$
    \frac{d}{dx} f(g(x)) = g'(x) f'(g(x)) 
$$

Теперь мы можем применить некоторые из приведенных выше правил, чтобы найти $u' = f'(x) = 3 \frac{d}{dx} x^2-4\frac{d}{dx}x = 6x-4$. Таким образом, устанавливая $x = 1$, мы имеем $u' = 2$: это подтверждается нашим более ранним экспериментом в этом разделе, где численный результат приближается к $2$. Эта производная также является наклоном касательной к кривой $u = f(x)$, когда $x = 1$.

## Частные производные

До сих пор мы имели дело с дифференцированием функций только одной переменной. При машинном обучении функции часто зависят от многих переменных. Таким образом, нам необходимо распространить идеи дифференцирования на эти многомерные функции. Пусть $y = f(x_1, x_2, \ldots, x_n)$ - функция с $n$ переменными. Частная производная $y$ по параметру $i^\mathrm{th}$ имеет вид

$$ \frac{\partial y}{\partial x_i} = \lim_{h \rightarrow 0} \frac{f(x_1, \ldots, x_{i-1}, x_i+h, x_{i+1}, \ldots, x_n) - f(x_1, \ldots, x_i, \ldots, x_n)}{h}.$$

В математическом анализе **частная производная** — одно из обобщений понятия производной на случай функции нескольких переменных. Частная производная — это предел отношения приращения функции по выбранной переменной к приращению этой переменной, при стремлении этого приращения к нулю.


Чтобы вычислить $\frac{\partial y}{\partial x_i}$, мы можем просто рассматривать $x_1, \ldots, x_{i-1}, x_{i+1}, \ldots, x_n$ как константы и вычислять производную из $y$ по $x_i$. Для обозначения частных производных следующие условия эквивалентны:

$$\frac{\partial y}{\partial x_i} = \frac{\partial f}{\partial x_i} = f_{x_i} = f_i = D_i f = D_{x_i} f.$$

Рассмотрим производную сложной функции:

Пусть имеются две функции $f(x)$ и $h(x)$, причем область значений $f$ принадлежит области определения
функции $h$. Тогда функция $h(f(x))$ является применением одной функции к результату другой и называется
сложной функцией.

Например, $f(x) = 1 + x$, а $h(x) = ln(x)$, тогда сложной функцией будет $(x) = h(f(x)) = ln(1 + x)$.

Дифференциалом $df$ функции $f(x)$ в точке $x_0$ называется линейная часть её приращения. Существует
связь между дифференциалом функции и её производной:

$$df = f'(x_0)dx$$
$$dx = \Delta x$$

Именно это выражение послужило основанием для использования знака дифференциала в обозначении Лейбница для производной:

$$f'(x_0) = \frac{df}{dx}(x_0)$$

Cледует особо отметить, что выражение $\frac{df}{dx}$ следует трактовать как единый символ, а не частное двух величин.

### Производная сложной функции

Рассмотрим вычисление производной сложной функции (Chain Rule), без доказательства.

Пусть $z = g(y)$, $y = h(x)$, тогда:

$$\frac{dg(h(x))}{dx} = \frac{dg(y)}{dy}\frac{dh(x)}{dx}$$

Теперь давайте посмотрим на более общий сценарий, в котором функции имеют произвольное количество переменных. Предположим, что дифференцируемая функция $y$ имеет переменные $u_1, u_2, \ldots, u_m$ где каждая дифференцируемая функция $u_i$ имеет переменные $x_1, x_2, \ldots, x_n$. Обратите внимание, что $y$ является функцией $x_1, x_2, \ldots, x_n$. Тогда по правилу вычисления производной сложной функции:

$$\frac{dy}{dx_i} = \frac{dy}{du_1} \frac{du_1}{dx_i} + \frac{dy}{du_2} \frac{du_2}{dx_i} + \cdots + \frac{dy}{du_m} \frac{du_m}{dx_i}$$

для любых $i = 1, 2, \ldots, n$.
